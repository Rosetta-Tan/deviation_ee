```python
import numpy as np
import os, sys
import json
out_dir = '/n/home01/ytan/scratch/deviation_ee/output_syk'
```

```python
Ls = [10, 15]
ents = []
ents_page = []
ents_avg = []
for L in Ls:
    sub_dir = '09112023_run_ee_eig_H2_gs_noparity/run_ee_eig_H2_gs_L{}/'.format(L)
    print('L:', L)
    ents.append([])
    for seed in range(10):
        filename = '000_{}.json'.format(seed)
        filepath = os.path.join(out_dir, sub_dir, filename)
        if L == 15 and seed in [0, 3]:
            continue
        else:
            with open(filepath, 'r') as f:
                data = json.load(f)
                ents[-1].append(data['sent'])
    ents_avg.append(np.mean(ents[-1]))
    ents_page.append(np.log(2) * L/2 - 0.5)
# ents_avg = np.mean(ents, axis=1)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(Ls, ents_avg, 'o-')
plt.plot(Ls, ents_page, 'o-')
plt.show()
```
